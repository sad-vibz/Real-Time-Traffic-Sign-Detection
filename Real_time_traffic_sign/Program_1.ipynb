```python
import cv2


class_names = ['-', '0', 'Barrier Ahead', 'Cattle', 'Caution', 'Cycle Crossing', 'Dangerous Dip', 'Eating Place', 'Falling Rocks', 'Ferry', 'First Aid Post', 'Give Way', 'Horn Prohibited', 'Hospital', 'Hump', 'Left Hair Pin Bend', 'Left Reverse Bend', 'Left hand curve', 'Light Refreshment', 'Men at Work', 'Narrow Bridge', 'Narrow road ahead', 'No Parking', 'No Stopping', 'No Thorough Road', 'No Thorough SideRoad', 'Parking Lot Cars', 'Parking Lot Cycle', 'Parking Lot Scooter and MotorCycle', 'Parking This side', 'Pedestrian Crossing', 'Pedestrian Prohibited', 'Petrol Pump- Gas Station', 'Public Telephone', 'Resting Place', 'Right Hair Pin Bend', 'Right Hand Curve', 'Right Reverse Bend', 'Road Wideness Ahead', 'Round About', 'School Ahead', 'Slippery Road', 'Speed Limit -10-', 'Speed Limit -100-', 'Speed Limit -110-', 'Speed Limit -120-', 'Speed Limit -130-', 'Speed Limit -140-', 'Speed Limit -150-', 'Speed Limit -160-', 'Speed Limit -20-', 'Speed Limit -25-', 'Speed Limit -35-', 'Speed Limit -45-', 'Speed Limit -48-', 'Speed Limit -5-', 'Speed Limit -50-', 'Speed Limit -55-', 'Speed Limit -60-', 'Speed Limit -65-', 'Speed Limit -70-', 'Speed Limit -75-', 'Speed Limit -8-', 'Speed Limit -80-', 'Speed Limit -90-', 'Speed Limit 3', 'Speed Limit 30', 'Speed limit -15-', 'Speed limit -40-', 'Steep Ascent', 'Steep Desecnt', 'Stop', 'Straight Prohibitor No Entry', 'walking']


cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

   
    results = model_1(frame)
    boxes = results[0].boxes  

    for box in boxes:
        cls = int(box.cls.item())         
        conf = box.conf.item()            
        x1, y1, x2, y2 = [int(coord.item()) for coord in box.xyxy[0]]  

        label_name = class_names[cls] if cls < len(class_names) else f"Class {cls}"
        label = f"{label_name}: {conf:.2f}"

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("Traffic Sign Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Start the webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break
    
    # Perform inference on the current frame
    results = model(frame)
    
    # Filter detections for traffic signs (class ID for stop sign, etc.)
    traffic_sign_classes = [9, 12]  # e.g., 9 = stop sign, 12 = speed limit sign (based on COCO dataset)

    filtered_results = []
    for *box, conf, cls in zip(*results.xywh[0]):
        if int(cls) in traffic_sign_classes:
            filtered_results.append((box, conf, cls))

    # Draw bounding boxes for filtered results
    for *box, conf, cls in filtered_results:
        x1, y1, x2, y2 = map(int, box)
        label = f"Traffic Sign: {conf:.2f}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(frame, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the frame with detections
    cv2.imshow("Traffic Sign Detection", frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import requests
import base64

# Replace with your info
ROBOFLOW_MODEL = "ati-yolov8-vv8qk/1"
API_KEY = "rf_iDQWNJBtBodewwVAXCHT7BTRXDC2"  # << paste your key here

# Build API URL
api_url = f"https://infer.roboflow.com/{ROBOFLOW_MODEL}?api_key={API_KEY}"

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Resize to model input size (Roboflow default is 416x416 or check your training size)
    resized = cv2.resize(frame, (416, 416))
    
    _, img_encoded = cv2.imencode('.jpg', resized)
    img_base64 = base64.b64encode(img_encoded).decode('utf-8')

    response = requests.post(
        api_url,
        data=img_base64,
        headers={"Content-Type": "application/x-www-form-urlencoded"}
    )

    detections = response.json().get("predictions", [])

    for pred in detections:
        class_name = pred["class"]
        confidence = pred["confidence"]
        x, y, width, height = int(pred["x"]), int(pred["y"]), int(pred["width"]), int(pred["height"])

        x1 = x - width // 2
        y1 = y - height // 2
        x2 = x + width // 2
        y2 = y + height // 2

        label = f"{class_name}: {confidence:.2f}"
        cv2.rectangle(resized, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(resized, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("Roboflow Detection", resized)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
